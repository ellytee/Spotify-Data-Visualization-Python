```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
df_track =pd.read_csv("tracks.csv")
df_track.head()
```

```python
#checking NA vallues
pd.isnull(df_track).sum()
```

```python
#checking least popular songs
sort_df= df_track.sort_values('popularity',ascending=True).head(5)
sort_df
```

```python
 df_track.describe().transpose()
```

```python
#making release date the index
df_track.set_index('release_date', inplace= True)
df_track.index=pd.to_datetime(df_track.index)
df_track.head(10)
```

```python
#changing duration to seconds from ms
df_track['duration']=df_track['duration_ms'].apply(lambda x: round(x/1000))
df_track.drop('duration_ms',inplace=True,axis=1)
df_track.duration.head()
```

```python
#correlation table
corr_df = df_track.drop(['key','mode','explicit'],axis=1).corr(method='pearson')
plt.figure(figsize=(14,6))
heatmap=sb.heatmap(corr_df ,vmin=-1 ,vmax=1 ,center=0 ,annot=True,fmt='.1g',cmap='inferno' ,linewidths=1 ,linecolor='white')
heatmap.set_title('Correlation between variables')
heatmap.set_xticklabels(heatmap.get_xticklabels(),rotation=90)
```
